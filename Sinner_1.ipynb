```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url = 'https://raw.githubusercontent.com/Ethan004-Code/Jannik_Sinner/main/jannik_sinner_tournament_summary_2025.xlsx'
df = pd.read_excel(url, engine='openpyxl')
```

```python

data = {
    'Tournament': ['French Open 2025', 'Wimbledon 2025'],
    'Matches Played': [5, 7],
    'Wins': [4, 6],
    'Losses': [1, 1],
    'Sets Won': [13, 18],
    'Sets Lost': [5, 6],
    'Aces': [45, 62],
    'Double Faults': [12, 15],
    'First Serve %': [65, 68],
    'Break Points Won': [22, 27],
    'Total Points Won': [430, 510],
    'Unforced Errors': [98, 85],
}

df = pd.DataFrame(data)
df.set_index('Tournament', inplace=True)
plt.figure(figsize=(12, 7))
df.T.plot(kind='bar', figsize=(14, 7), width=0.7)
plt.title("Jannik Sinner Performance Comparison: French Open vs Wimbledon (2025)")
plt.ylabel("Value")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.legend(title='Tournament')
plt.show()
```

```python

data = {
    'Metric': ['Matches Played', 'Wins', 'Losses', 'Sets Won', 'Sets Lost',
               'Aces', 'Double Faults', 'First Serve %', 'Break Points Won',
               'Total Points Won', 'Unforced Errors'],
    'French Open': [5, 4, 1, 13, 5, 45, 12, 65, 22, 430, 98],
    'Wimbledon': [7, 6, 1, 18, 6, 62, 15, 68, 27, 510, 85]
}

df = pd.DataFrame(data)
df['Difference'] = df['Wimbledon'] - df['French Open']
df = df.sort_values('Difference', key=abs, ascending=True)

plt.figure(figsize=(10, 7))
colors = df['Difference'].apply(lambda x: 'green' if x > 0 else 'red')

plt.barh(df['Metric'], df['Difference'], color=colors)
plt.axvline(0, color='gray', linestyle='--')
plt.title("Jannik Sinner: Wimbledon vs French Open (2025)\nDiverging Bar Chart")
plt.xlabel("Positive = Wimbledon Better | Negative = French Open Better")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
